```python
A_2 = sp.Matrix([[2,1,0],[2,2,0],[0,0,1]]);A_2
```

```python
B_2 = sp.Matrix([[1,0,0],[0,-1,0],[0,0,2]]);B_2
```

```python
# C = (2A-I)B(2A-I)^{-1}
C_2 = (2*A_2-sp.eye(3))*B_2*((2*A_2-sp.eye(3)).inv());C_2
```

```python
(2*C_2*A_2 - 2*A_2*B_2) == (C_2 - B_2)
```
